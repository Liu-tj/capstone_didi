## 9 AM Analysis

```python
taxi_weekday['time'] = pd.to_datetime('2019-01-01 09:00:00')
```

```python
taxi_weekday.head()
```

```python
taxi_weekday.tail()
```

```python
taxi_weekday['diff'] = (taxi_weekday['time'] - taxi_weekday['pickup_datetime']).dt.seconds / 3600
```

```python
final = taxi_weekday[taxi_weekday['diff'] < 1]
```

```python
final.head()
```

```python
pickups = pd.DataFrame(final.PULocationID.value_counts()).reset_index()
pickups.columns = ['PULocationID', 'counts']
```

```python
pickups = pd.merge(pickups, lookup, left_on='PULocationID', right_on='LocationID', how='left')
```

```python
pickups[pickups.Borough=='Manhattan']
```
